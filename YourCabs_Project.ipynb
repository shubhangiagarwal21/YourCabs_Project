```python
cleaned_data=data.copy()
```

```python
# cleaned_data['cancl_density_from_area'].value_counts()
```

```python
# data_merged = pd.concat([data,target],axis=1,join='inner')
cleaned_data['from_area_id'] = round(cleaned_data.groupby('from_area_id')['Car_Cancellation'].sum()/cleaned_data.groupby('from_area_id')['Car_Cancellation'].count(),2)
cleaned_data['from_area_id'].replace(np.nan,0,inplace=True)
```

```python
cond = [(cleaned_data['from_area_id'].astype('float').between(0,0.33)),
        (cleaned_data['from_area_id'].astype('float').between(0.34,0.66)),
        (cleaned_data['from_area_id'].astype('float').between(0.67,1.0))]
values = ['Low Cancellation','Medium Cancellation','High Cancellation']
cleaned_data['from_area_id'] = np.select(cond,values)
```

```python
cleaned_data['to_area_id'] = round(cleaned_data.groupby('to_area_id')['Car_Cancellation'].sum()/cleaned_data.groupby('to_area_id')['Car_Cancellation'].count(),2)
cleaned_data['to_area_id'].replace(np.nan,0,inplace=True)
```

```python
cond = [(cleaned_data['to_area_id'].astype('float').between(0,0.33)),
        (cleaned_data['to_area_id'].astype('float').between(0.34,0.66)),
        (cleaned_data['to_area_id'].astype('float').between(0.67,1.0))]
values = ['Low Cancellation','Medium Cancellation','High Cancellation']
cleaned_data['to_area_id'] = np.select(cond,values)
```

```python
def cal_distance(from_lat,from_long,to_lat,to_long):
    return distance.distance((from_lat,from_long),(to_lat,to_long)).km
```

```python
cleaned_data['distance'] = cleaned_data.apply(lambda row: cal_distance(row['from_lat'],row['from_long'],row['to_lat'],row['to_long']),axis=1)
```

```python
cleaned_data['from_date_day'] = pd.to_datetime(cleaned_data['from_date_dt']).dt.day_name()
cleaned_data['booking_created_day'] = pd.to_datetime(cleaned_data['booking_created_dt']).dt.day_name()
cleaned_data['from_date_month'] = pd.to_datetime(cleaned_data['from_date_dt']).dt.month_name()
cleaned_data['booking_created_month'] = pd.to_datetime(cleaned_data['booking_created_dt']).dt.month_name()
cleaned_data['from_date_week'] = np.where((cleaned_data['from_date_day']=='Saturday') | (cleaned_data['from_date_day']=='Sunday'),'Weekend','Weekday',)
cleaned_data['booking_created_week'] = np.where((cleaned_data['booking_created_day']=='Saturday') | (cleaned_data['booking_created_day']=='Sunday'),'Weekend','Weekday',)
```

```python
cond = [(pd.to_datetime(cleaned_data['from_time_tm']).dt.hour.between(5, 8)),
        (pd.to_datetime(cleaned_data['from_time_tm']).dt.hour.between(9, 12)),
        (pd.to_datetime(cleaned_data['from_time_tm']).dt.hour.between(13, 16)),
        (pd.to_datetime(cleaned_data['from_time_tm']).dt.hour.between(17, 20)),
        ((pd.to_datetime(cleaned_data['from_time_tm']).dt.hour.between(21, 24)) | (pd.to_datetime(cleaned_data['from_time_tm']).dt.hour==0)),
        (pd.to_datetime(cleaned_data['from_time_tm']).dt.hour.between(1, 4))]
values = ['Early Morning','Morning','Afternoon','Evening','Night','Late Night']
cleaned_data['from_date_session'] = np.select(cond,values)
```

# Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
num = cleaned_data.select_dtypes(include='number')
char = cleaned_data.select_dtypes(include='object')
```

```python
X_char= pd.get_dummies(char, drop_first=True)
X_char
```

```python
num.head()
```

```python
data_all = pd.concat([num,X_char],axis=1,join='inner')
```

```python
X = data_all.drop('Car_Cancellation',axis=1)
y = data_all['Car_Cancellation']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state= 131)
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train, y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision",metrics.precision_score(y_test,y_pred))
print("Recall",metrics.recall_score(y_test,y_pred))
print("f1_score",metrics.f1_score(y_test,y_pred))
```
